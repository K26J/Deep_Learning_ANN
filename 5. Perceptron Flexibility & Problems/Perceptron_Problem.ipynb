```python
from sklearn.linear_model import Perceptron
import numpy as np
```

```python
# XOR Data
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 1, 1, 0])

# Initialize Perceptron
clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(X, y)

# Try to predict
print("Predictions on XOR data:", clf.predict(X))
print("Accuracy:", clf.score(X, y))

# Output will usually be 50% (0.5), proving it fails to learn the pattern.
# It usually predicts all 0s or a mix that gets half wrong.
```
