```python
def forward(self, X):
        # --- Layer 1 ---
        # Z1 is the linear combination
        self.Z1 = np.dot(X, self.W1) + self.b1
        # A1 is the activation (Output of hidden layer)
        self.A1 = self.sigmoid(self.Z1)
        
        # --- Layer 2 ---
        # Input for this layer is A1 (output of previous layer)
        self.Z2 = np.dot(self.A1, self.W2) + self.b2
        # Final prediction
        self.A2 = self.sigmoid(self.Z2)
        
        return self.A2
```
