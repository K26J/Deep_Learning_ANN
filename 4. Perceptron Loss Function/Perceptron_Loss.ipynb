```python
import numpy as np
```

```python

def perceptron_loss(X, y, weights, bias):
    """
    Calculates the total loss for a Perceptron.
    y should be -1 or 1 for this formula.
    """
    total_loss = 0
    for i in range(len(X)):
        # Calculate the linear output z = w*x + b
        prediction = np.dot(X[i], weights) + bias
        
        # Check misclassification
        # If y*prediction is negative, the signs are different (WRONG)
        if y[i] * prediction <= 0:
            # Loss is the absolute value of the prediction (distance score)
            total_loss += -y[i] * prediction
            
    return total_loss

```

```python
# Example Usage
X_sample = np.array([[2, 3], [1, 1], [4, 5]])
y_sample = np.array([1, -1, 1]) 
w_sample = np.array([0.1, 0.2])
b_sample = 0.5

loss = perceptron_loss(X_sample, y_sample, w_sample, b_sample)
print(f"Current Total Loss: {loss}")
```
