```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def plot_decision_boundary(X, y, weights, bias):
    # Plotting the data points
    plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', s=100)
    
    # Calculating the decision boundary line
    # The line equation is: w1*x1 + w2*x2 + b = 0
    # So, x2 = -(w1*x1 + b) / w2
    x1_min, x1_max = X[:,0].min() - 1, X[:,0].max() + 1
    x1_values = np.linspace(x1_min, x1_max, 100)
    x2_values = -(weights[0] * x1_values + bias) / weights[1]

    plt.plot(x1_values, x2_values, 'k--', label='Decision Boundary')
    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 1.5)
    plt.legend()
    plt.title("Perceptron Decision Boundary")
    plt.show()
```

```python
if __name__ == "__main__":
    # Using the weights we hypothetically learned in step 2
    # Example weights for AND gate: w=[0.1, 0.2], b=-0.2 (Just for demo)
    X = np.array([[0,0], [0,1], [1,0], [1,1]])
    y = np.array([0, 0, 0, 1])
    
    # You can plug in the weights you got from Folder 2 here
    demo_weights = np.array([0.1, 0.1]) 
    demo_bias = -0.15 
    
    plot_decision_boundary(X, y, demo_weights, demo_bias)
```
